# Superbowl History


```python
# Import Data Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Imported Superbowl Dataset
df = pd.read_csv(r"C:\Users\Issa.Flynn\Desktop\superbowl.csv")
df.dropna(axis=1)
df.fillna('NaN')
df.drop_duplicates()
df.info()
```

```python
#Printed out the first 5 rows of our data
df.head()
```

```python
#Converted the data into a dataframe
pd.DataFrame(df)
```

```python
df.describe()
```

```python
#importing csv with superbowl records from 2021 onwards
df2 = pd.read_csv (r"C:\Users\Issa.Flynn\OneDrive - Bytedance\superbowl_new.csv")
df2
```

```python
# Merge the 2 dataframes together
combined = ([df, df2])
superbowl = pd.concat(combined)
superbowl
```

```python
#sorted the values by year in descending order and set the date as the index
superbowl.sort_values('Year', ascending=False).set_index('Date')
```

Now that I have my data in order, I am going to see who has won/lost the superbowl on the most occasions and represent this on 2 seperate bar charts.

```python
# Winners bar plot
superbowl['Times'] = 1
superbowl_grp_win = superbowl.groupby('Winner').sum()['Times'].reset_index().sort_values(by = 'Times', ascending=False)
superbowl_grp_win
```

```python
plt.figure(figsize=(68,18))
sns.barplot(x='Winner', y='Times', data=superbowl_grp_win)
```

```python
# Losers bar plot
superbowl_grp_los = superbowl.groupby('Loser').sum()['Times'].reset_index().sort_values(by = 'Times', ascending=False)
superbowl_grp_los
```

```python
plt.figure(figsize=(68,18))
sns.barplot(x='Loser', y='Times', data=superbowl_grp_los)
```
